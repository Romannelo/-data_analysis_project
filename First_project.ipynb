```python
#меня столбцу ЭТО АПАРТЫ на булиновый тип, чтобы в будущем по нему было проще рабоать, если понадобится
data['is_apartment'] = data['is_apartment'].astype(bool)
```

```python
# Смотри уникальные значения в названия нас. пунктов, чтобы убрать неявные дубликаты
data['locality_name'].sort_values().unique()
```

```python
#собственно, их же писали люди, и не будем создавать путаницу в отчетах, убереим ее
data['locality_name']=data['locality_name'].str.replace('посёлок','поселок')
data['locality_name']=data['locality_name'].str.replace('поселок городского типа','поселок')
```

```python
#проверяю, 
data['locality_name'].sort_values().unique()
```

```python
living_zone_median = data.groupby('rooms')['living_area'].median()
for room in data['rooms'].unique():
    data.loc[((data['living_area'].isna()) & (data['rooms'] == room)), 'living_area'] = living_zone_median[room]

kitchen_zone_median = data.groupby('rooms')['kitchen_area'].median()
for room in data['rooms'].unique():
    data.loc[((data['kitchen_area'].isna()) & (data['rooms'] == room)), 'kitchen_area'] = kitchen_zone_median[room]
```

# Шаг 3. Добавьте в таблицу новые столбцы со следующими параметрами:
* цена одного квадратного метра (нужно поделить стоимость объекта на его общую площадь, а затем округлить до двух знаков после запятой);
* день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
* месяц публикации объявления;
* год публикации объявления;
* тип этажа квартиры (значения — «‎первый», «последний», «другой»);
* расстояние до центра города в километрах (переведите из м в км и округлите до ближайших целых значений).

```python
data['1m_price'] = round(data['last_price']/data['total_area'],2)  #считаю цену метра квадратного
#перевожу 'first_day_exposition' в формат дэйттайм
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])
#Ниже вытягиваю день, месяц и год в отдельные столбцы
data['days_num'] = data['first_day_exposition'].dt.weekday
data['mounths_num'] = data['first_day_exposition'].dt.month
data['years_num'] = data['first_day_exposition'].dt.year
```

```python
#Сдлеаем тип этажа
def floor_type(df):
    if df['floor'] == 1:
        return 'первый'
    elif df['floor'] == df['floors_total']:
        return 'последний'
    else:
        return 'другой'
#Лучший курс яндекс практикума научил меня применять функцию ко всему датафрейму сразу и причем методом аксис указываем, что будем брать столбцы, а не строки)
data['floor_type'] = data.apply(floor_type,axis=1)
```

```python
#Посчитаем расстояние до центра в км и оруглим до ближ. целого

data['cityCenters_nearest_on_KM'] = round(data['cityCenters_nearest']/1000,0)
data
```
